```python
import doctest

class ShoppingCart:

    def __init__(self):
        self.myStoreInventory = {
            'Apparel': {
                'Jersey': 60.0,
                'Shorts': 30.0,
                'Socks': 10.0
            },
            'Shoes': {
                'Running Shoes': 120.0,
                'Basketball Shoes': 150.0,
                'Casual Shoes': 80.0
            },
            'Balls': {
                'Basketball': 25.0,
                'Football': 30.0,
                'Baseball': 15.0,
                'Golf ball': 5.0,
                'Soccer ball': 20.0
            }
        }
        self.cart = []

    def add_item(self, category, item):
        """
        Add an item to the shopping cart. Each item belongs to a category and has a distinct name.

        >>> cart = ShoppingCart()
        >>> cart.add_item('Apparel', 'Jersey')
        >>> cart.cart
        [('Apparel', 'Jersey', 60.0)]
        """
        self.cart.append((category, item, self.myStoreInventory[category][item]))

    def total_due(self):
        """
        Subtotal for all items in the cart inclusive of 8% tax.

        >>> cart = ShoppingCart()
        >>> cart.add_item('Apparel', 'Jersey')
        >>> cart.add_item('Balls', 'Basketball')
        >>> f"{cart.total_due():.2f}"
        '91.80'
        """
        subtotal = sum(price for _, _, price in self.cart)
        tax = subtotal * 0.08
        subtotal=subtotal + tax
        return subtotal
```

```python
if __name__ == "__main__":

    mycart=ShoppingCart()
    mycart.add_item('Apparel', 'Jersey')
    mycart.add_item('Balls', 'Basketball')
    print("Total for all items in the cart with tax is: $" + f"{mycart.total_due():.2f}")
    doctest.testmod()
```
