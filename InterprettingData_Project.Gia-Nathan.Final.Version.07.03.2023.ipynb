```
################ 1. Normal distribution using scipy.stats ################
#   The scipy.stats norm function can be used to generate samples
#   according to the normal (gaussian) distribution.
#   https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html
#   Generate a dataset of x_axis values using numpy arange from -5 to +5 in
#   increments of 0.001.  Generate a dataset of y_axis values using the
#   scipy.stats norm function with mean 0 and standard deviation 1.
#   Plot the distribution using matplotlib pyplot.
#   Upload a screenshot of the distribution plot to the assignment doc in the LMS.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline


#Generate a dataset of x_axis values using numpy arange from -5 to +5 in increments of 0.001
#Your code here...
#x_axis = ...
#y_axis = ...
x_axis = np.arange(-5, 5, 0.001)
y_axis = stats.norm.pdf(x_axis, loc=0, scale=1)
#Plot the distribution using matplotlib pyplot
#Create the title 'Normal Distribution'
plt.plot(x_axis, y_axis)
plt.title('Normal Distribution')
plt.show()



################ 2. Normal distribution using numpy.randn #################
#   The numpy randn library can be used to generate random
#   samples from the normal distribution:
#   https://numpy.org/doc/stable/reference/random/generated/numpy.random.randn.html
#   Use numpy.random randn function to generate 10,000 samples centered around 0
#   and plot the distribution on a histogram plot with 10 bins and the same data
#   with 50 bins (plot both distributions on the same histogram).
#   Upload a screenshot of the histogram plot to the assignment doc in the LMS.

from numpy.random import seed
from numpy.random import randn, randint

#Use numpy.random randn function to generate 10,000
#samples centered around 0 and plot the distribution
#on a histogram plot with 10 bins and the same data
#with 50 bins (plot both distributions on the same histogram).
#####Your code here...

# Set the seed for reproducibility (optional)
np.random.seed(42)

# Generate 10,000 samples from a normal distribution centered around 0
data = np.random.randn(10000)

# Create a histogram plot with 10 bins
plt.hist(data, bins=10, alpha=0.5, label='10 bins')

# Create a histogram plot with 50 bins
plt.hist(data, bins=50, alpha=0.5, label='50 bins')

# Add labels and a legend to the plot
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Display the plot
plt.show()


################ 3. Numpy mean, median, standard deviation ################
#   The numpy mean, median, std functions can used to calculate SOCV statistics
#   Create 10,000 randn samples with a random offset that changes each
#   time you run the cell.  e.g.  data = randn(10000) + randint(1,50)
#   Calculate and print the mean, median and standard deviation of the distribution.
#   Record the mean, median, standard deviation for 10 runs of the cell.
#   How does each of the statistics change?  Upload a screenshot of your
#   results (mean, median, standard deviation) to the assignment doc in the LMS.
#   Comment in the assignment doc in the LMS on which of the statistics changed and why?

from numpy.random import seed
from numpy.random import randn
from numpy import mean, median, std
import random

#Create 10,000 randn samples with a random
#offset that changes each time you run the cell.
#####Your code here...
from numpy.random import seed
from numpy.random import randn, randint
from numpy import mean, median, std

# Set the seed for reproducibility (optional)
seed(42)

# Create 10,000 randn samples with a random offset
data = randn(10000) + randint(1, 50)

# Calculate the mean, median, and standard deviation of the distribution
mean_val = mean(data)
median_val = median(data)
std_val = std(data)

# Print the mean, median, and standard deviation
print("Mean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_val)


################ 4. Interquartile Range ################
#   The numpy percentile and scipy.stats iqr functions can be used to
#   determine the interquartile range and identify data outliers:
#   https://numpy.org/doc/stable/reference/generated/numpy.percentile.html
#   https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.iqr.html
#   Note:  data4 and data4b have been defined below in the doctest module.
#   Define a function which calculates and prints the first and third quartile,
#   then find and print the interquartile range (IQR).  Use the doctest module to test your function.
#   Define a function which calculates and prints the IQR, upper outlier threshold,
#   and lower outlier threshold.
#   Define a function which calculates and prints any outliers from the dataset.
#   Use the doctest module to test your function
#   Upload screenshots of your results and the output from the doctest module to the assignment doc in the LMS.


#Calculate the Q1, Q3, and IQR
#def IQR_calculator(dataIn):
  #dataIn = list of data
  #return IQR, Q1, Q3

  #Calculate first and third quartiles (Q1, Q3) using numpy percentile function
  #Or calculate IQR directly using scipy.stats iqr
  #####Your code here...

  # Calculate the Interquartile range from numpy function
  #####Your code here...


#Calculate the IQR outlier thresholds
#def IQR_outlier_thresholds(IQR, Q1, Q3):
  #IQR = interquartile range
  #Q1, Q3 = Quartile 1, 3
  #return tuple of low and high outlier thresholds

  #####Your code here...


#Find the outliers
#def IQR_find_outliers(dataIn, lowT, highT):
  #dataIn = list of data
  #highT = high outlier threshold
  #lowT = low outlier threshold
  #return list of low outliers, list of high outliers (outlierLow, outlierHigh)

  #####Your code here...
# Calculate the Q1, Q3, and IQR
def IQR_calculator(dataIn):
    # Calculate first quartile (Q1)
    q1 = np.percentile(dataIn, 25)

    # Calculate third quartile (Q3)
    q3 = np.percentile(dataIn, 75)

    # Calculate interquartile range (IQR)
    iqr = q3 - q1

    return iqr, q1, q3

# Calculate the IQR outlier thresholds
def IQR_outlier_thresholds(IQR, Q1, Q3):
    # Calculate lower outlier threshold
    low_thresh = Q1 - 1.5 * IQR

    # Calculate upper outlier threshold
    high_thresh = Q3 + 1.5 * IQR

    return low_thresh, high_thresh

# Find the outliers
def IQR_find_outliers(dataIn, lowT, highT):
    # Find the outliers below the lower threshold
    low_outliers = dataIn[dataIn < lowT]

    # Find the outliers above the upper threshold
    high_outliers = dataIn[dataIn > highT]

    return low_outliers, high_outliers




################ 5. Create boxplots ################
#   Create boxplots for the 2 datasets in #4.
#   Upload screenshots of your plots to the assignment doc in the LMS.
#   Confirm the medan, Q1, Q3, and outlier points on the plots.
#def plot_results():
  #no input
  #no return
  #####Your code here...

def plot_results():
    # Create a figure with two subplots
    fig, axs = plt.subplots(1, 2, figsize=(10, 6))

    # Plot the boxplot for data4
    axs[0].boxplot(data4, vert=False)
    axs[0].set_title('data4 Boxplot')

    # Plot the boxplot for data4b
    axs[1].boxplot(data4b, vert=False)
    axs[1].set_title('data4b Boxplot')

    # Add labels to the y-axis
    axs[0].set_yticklabels(['data4'])
    axs[1].set_yticklabels(['data4b'])

    # Adjust the spacing between subplots
    plt.subplots_adjust(wspace=0.5)

    # Display the plot
    plt.show()


#Verbose print function
def verbosePrint():
        print('IQR_calculator (IQR - data4):', iqr)
        print('IQR_calculator (Q1 - data4):', q1)
        print('IQR_calculator (Q3 - data4):', q3)

        print('IQR_outlier_thresholds (lowThresh - data4):', iqrOutlierLowT)
        print('IQR_outlier_thresholds (highThresh - data4):', iqrOutlierHighT)

        print('IQR_find_outliers (low outliers - data4):', lowOL)
        print('IQR_find_outliers (high outliers - data4):', highOL)

        print('IQR_calculator (IQR - data4b):', iqr_b)
        print('IQR_calculator (Q1 - data4b):', q1_b)
        print('IQR_calculator (Q3 - data4b):', q3_b)

        print('IQR_outlier_thresholds (lowThresh - data4b):', iqrOutlierLowT_b)
        print('IQR_outlier_thresholds (highThresh - data4b):', iqrOutlierHighT_b)

        print('IQR_find_outliers (low outliers - data4b):', lowOL_b)
        print('IQR_find_outliers (high outliers - data4b):', highOL_b)




#-------------------------------------------------------------------------------------------------
#Test with the following doctest test vectors.
#DO NOT EDIT THE TEST CODE!!!!
#Even changing the spacing can cause errors.
#The test code will automatically execute when you run the cell.
#You should test all your combination of outputs but your code at least must pass these exact tests.
#If your code fails, you will see a description in the console cell.
#If your code passes, you will see the message: "TestResults(failed=0, attempted=X)" where X is the number of tests.
import doctest

data4 = np.array([6, 5, 7, 22, 6, 5, 6, 8, 9, 12, 10, 20, 2, 35, 7, 2, 13, 15, 4, 10])
data4b = np.array([32, 88, 36, 46, 92, 56, 101, 75, 79, 79, 1, 89, 91, 47, 116, 96, 97, 69, 170, 112, 93, 105])

iqr,q1,q3 = IQR_calculator(data4)
iqr_b,q1_b,q3_b = IQR_calculator(data4b)
iqrOutlierLowT, iqrOutlierHighT = IQR_outlier_thresholds(iqr, q1, q3)
iqrOutlierLowT_b, iqrOutlierHighT_b = IQR_outlier_thresholds(iqr_b, q1_b, q3_b)
lowOL, highOL = IQR_find_outliers(data4, iqrOutlierLowT, iqrOutlierHighT)
lowOL_b, highOL_b = IQR_find_outliers(data4b, iqrOutlierLowT_b, iqrOutlierHighT_b)
verbosePrint()
plot_results()

"""
  >>> print(iqr,q1,q3)
  6.5 5.75 12.25
  >>> print(iqrOutlierLowT,iqrOutlierHighT)
  -4.0 22.0
  >>> print(lowOL,highOL)
  [] [35]
  >>> print(lowOL_b,highOL_b)
  [1] [170]
"""

doctest.testmod()
```
