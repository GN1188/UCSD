```python
import doctest

def main():
    interview_questions = ["Name", "Age", "GPA"]
    interview_responses = []

    for i in range(3):
        question = interview_questions[i]
        response = input("Enter {}:".format(question))
        interview_responses.append(response)

    employees = []
    for i in range(5):
        employee = []
        for j in range(3):
            employee.append(input("Enter {} for Employee{}:".format(interview_questions[j], i+1)))
        employees.append(employee)

    print("Categories of Employees:")
    for question in interview_questions:
        print(question)

    category_of_interest = int(input("Enter the category of interest (0 for Name, 1 for Age, 2 for GPA):"))

    print("Manager's category of interest for each employee:")
    for employee in employees:
        print(employee[category_of_interest])


if __name__ == "__main__":
    main()


```
