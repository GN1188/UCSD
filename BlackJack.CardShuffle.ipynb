```python
def perfect_shuffle(deck):
    length = len(deck)
    shuffled_deck = [0] * length
    index = 0

    # Interleave the first half of the deck
    for i in range(length // 2):
        shuffled_deck[index] = deck[i]
        index += 2

    index = 1

    # Interleave the second half of the deck
    for i in range(length // 2, length):
        shuffled_deck[index] = deck[i]
        index += 2

    return shuffled_deck
```

```python
my_deck = list(range(52))  # Assuming a standard deck of 52 cards
shuffled_deck = perfect_shuffle(my_deck)
print(shuffled_deck)
```

```python
import random

def random_shuffle(deck, num_shuffles):
    length = len(deck)

    for _ in range(num_shuffles):
        for i in range(length):
            # Generate a random index within the range of the deck
            j = random.randint(0, length - 1)

            # Swap the cards at indices i and j
            deck[i], deck[j] = deck[j], deck[i]

    return deck
```

```python
my_deck = list(range(52))  # Assuming a standard deck of 52 cards
num_shuffles = int(input("Enter the number of shuffles: "))
shuffled_deck = random_shuffle(my_deck, num_shuffles)
print(shuffled_deck)
```

```python
import random

class Card:
    def __init__(self, rank, suit):
        self.rank = rank
        self.suit = suit

    def __str__(self):
        return f"{self.rank} of {self.suit}"

class Deck:
    def __init__(self):
        ranks = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']
        suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
        self.cards = [Card(rank, suit) for rank in ranks for suit in suits]

    def shuffle(self):
        random.shuffle(self.cards)

    def deal_card(self):
        return self.cards.pop()

def deal(deck):
    player_hand = [deck.deal_card(), deck.deal_card()]
    dealer_hand = [deck.deal_card(), deck.deal_card()]
    return player_hand, dealer_hand

def hit(hand, deck):
    hand.append(deck.deal_card())

def calculate_score(hand):
    score = sum(get_card_value(card) for card in hand)
    if 'Ace' in [card.rank for card in hand] and score <= 11:
        score += 10  # Treat Ace as 11 if it doesn't bust
    return score

def get_card_value(card):
    if card.rank in ['Jack', 'Queen', 'King']:
        return 10
    elif card.rank == 'Ace':
        return 1
    else:
        return int(card.rank)

def play_blackjack():
    print("Let's play Blackjack!")
    deck = Deck()
    deck.shuffle()

    while True:
        player_hand, dealer_hand = deal(deck)
        player_score = calculate_score(player_hand)
        dealer_score = calculate_score(dealer_hand)

        print(f"Player's hand: {[str(card) for card in player_hand]}")
        print(f"Dealer's hand: [{str(dealer_hand[0])}, ?]")

        if player_score == 21 and len(player_hand) == 2:
            print("Player wins with a Blackjack!")
        elif dealer_score == 21 and len(dealer_hand) == 2:
            print("Dealer wins with a Blackjack!")
        else:
            while player_score < 21:
                action = input("Do you want to hit or stay? ").lower()
                if action == "hit":
                    hit(player_hand, deck)
                    player_score = calculate_score(player_hand)
                    print(f"Player's hand: {[str(card) for card in player_hand]}")
                elif action == "stay":
                    break

            while dealer_score < 16:
                hit(dealer_hand, deck)
                dealer_score = calculate_score(dealer_hand)

            print(f"\nPlayer's hand: {[str(card) for card in player_hand]}")
            print(f"Dealer's hand: {[str(card) for card in dealer_hand]}")

            if player_score > 21:
                print("Player busts! Dealer wins.")
            elif dealer_score > 21:
                print("Dealer busts! Player wins.")
            elif player_score > dealer_score:
                print("Player wins!")
            elif dealer_score > player_score:
                print("Dealer wins!")
            else:
                print("It's a tie!")

        play_again = input("Do you want to play again? (yes/no) ").lower()
        if play_again != "yes":
            break

play_blackjack()
```
