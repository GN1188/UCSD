```python
from collections import deque

class ToDoTasksQueue:
    def __init__(self):
        self.tasks = deque()

    def add_task(self, task):
        """
        Add a task to the queue.
        """
        self.tasks.append(task)
        return f"Task {task} added!"

    def first_task(self):
        """
        Display the first task in the queue.
        """
        if not self.tasks:
            return "No tasks in the list!"
        return f"Your first task is {self.tasks[0]}"

    def complete_task(self):
        """
        Complete and remove the first task in the queue.
        """
        if not self.tasks:
            return "No tasks in the list!"
        return f"You removed the task {self.tasks.popleft()}"

    def clear_tasks(self):
        """
        Removes all the tasks in the queue.
        """
        self.tasks.clear()
        return "All tasks are cleared!"

def main():
    tasks = ToDoTasksQueue()
    while True:
        print("Welcome to my queue task list maker!")
        print("What would you like to do today?")
        print("1: Add a task")
        print("2: Look at the first task")
        print("3: Complete (remove) the first task")
        print("4: Clear the tasks")
        print("5: Exit")
        option = int(input())
        if option == 1:
            task = input("Add your task:\n")
            print(tasks.add_task(task))
        elif option == 2:
            print(tasks.first_task())
        elif option == 3:
            print(tasks.complete_task())
        elif option == 4:
            print(tasks.clear_tasks())
        elif option == 5:
            print("Goodbye!")
            break

if __name__ == "__main__":
    main()
```
