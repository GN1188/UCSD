```python
import doctest
def ageCategory2(age):
    if age < 5:
        message = "You belong to the age category: Less than 5"
    elif age >= 5 and age <= 10:
        message = "You belong to the age category: Between 5 and 10 inclusive"
    elif age >= 11 and age <= 20:
        message = "You belong to the age category: Between 11 and 20 inclusive"
    elif age >= 21 and age <= 64:
        message = "You belong to the age category: Between 21 and 64 inclusive"
    else:
        message = "You belong to the age category: Greater than 64"

    #print(message)
    return message
"""
Test the following function:
#ageCategory2(age):

>>> ageCategory2(15)
'You belong to the age category: Between 11 and 20 inclusive'
"""

# Call the function from the main program
if __name__ == "__main__":
    age = int(input("Enter your age: "))
    result = ageCategory2(age)
    print("Returned output: ", result)
    doctest.testmod()

```
