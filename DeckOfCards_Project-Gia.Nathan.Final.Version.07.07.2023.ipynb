Blackjack - Create the game with a dealer and single player with the following requirements:
* Your game must contain classes for the cards and deck
* Must contain a shuffle function to randomly shuffle the cards in the deck
* Must contain a deal function to deal the cards to the dealer and player
* Must contain a 'hit' or 'stay' function for the player and dealer (user input for player)
* Must display the winner, whether the dealer or player busted (went over 21), and prompt the player to play again.  If playing again, the cards in the deck are replaced and reshuffled.
* Player hits/stays first then dealer stays if >= 16, else hits until >= 16 or busted
* Must have logic to treat Ace as 1 or 11 (use 1 if busted)
* If a user or dealer is dealt a Jack and Ace then they win and game is over


```python
#Blackjack Game with simple implementation


from random import randint
from copy import deepcopy

# Classes for Card and Deck are similar as in Part 1
# Added some shortcuts for


class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank

class Deck:
    def __init__(self):

        # modified the initialization from the class above to make it simpler for debugging

        self.cards = [Card(suit, rank) for suit in ["Hearts", "Diamonds", "Clubs", "Spades"] for rank in range(1, 14)]

    def perfectShuffle(self):
        mid = len(self.cards) // 2
        shuffDeckPerf = [0] * len(self.cards)
        shuffDeckPerf[::2] = self.cards[:mid]
        shuffDeckPerf[1::2] = self.cards[mid:]
        self.cards = shuffDeckPerf

        # modified this function, as the user input for number of shuffles is in the deal function.

    def randomShuffle(self, num_shuffles):

        #shuffle function to randomly shuffle the cards in the deck

        for _ in range(num_shuffles):
            for i in range(len(self.cards)-1, 0, -1):
                j = randint(0, i)
                self.cards[i], self.cards[j] = self.cards[j], self.cards[i]

class BlackJack:

    def __init__(self):
        self.deck = Deck()
        self.player_hand = []
        self.dealer_hand = []

    def deal(self):

         #deal function to deal the cards to the dealer and player

        self.player_hand.append(self.deck.cards.pop())
        self.dealer_hand.append(self.deck.cards.pop())
        self.player_hand.append(self.deck.cards.pop())
        self.dealer_hand.append(self.deck.cards.pop())

    def hit(self, hand):
        hand.append(self.deck.cards.pop())

    def calculate_hand(self, hand):
        ace = 0
        total = 0
        for card in hand:
            if card.rank > 10:
                total += 10
            elif card.rank == 1:
                ace += 1
                total += 11
            else:
                total += card.rank
        while total > 21 and ace:
            total -= 10
            ace -= 1
        return total

    def play(self):
        self.deck.randomShuffle(int(input("Enter the number of shuffles: ")))
        self.deal()

        while True:
            player_total = self.calculate_hand(self.player_hand)
            dealer_total = self.calculate_hand(self.dealer_hand)
            print(f"Your hand is as follows: {[(card.suit, card.rank) for card in self.player_hand]}, total: {player_total}")
            print(f"Dealer's hand is as follows: {[(card.suit, card.rank) for card in self.dealer_hand]}, total: {dealer_total}")

            if player_total > 21:
                print("Player You are busted. So Dealer wins.")
                return
            elif dealer_total > 21:
                print("Dealer is busted. So Player you win.")
                return
            elif 21 >= player_total >= 17 and dealer_total >= 17:
                if player_total > dealer_total:
                    print("You win!")
                    return
                else:
                    print("Dealer wins.")
                    return

            action = input("Hit or stay? ").lower()
            if action == 'hit':
                self.hit(self.player_hand)
            if dealer_total < 16:
                self.hit(self.dealer_hand)

blackjack = BlackJack()
blackjack.play()
```
