```python
# function for interviews and the responses
def interview():
    name = input("Please enter your name: ")
    age = input("Please enter your age: ")
    gpa = input("Please enter your GPA: ")
    return [name, age, gpa]

# function to prompt manager for category of interest and print
def print_category(employees, category):
    categories = {0: "Name", 1: "Age", 2: "GPA"}
    print(f"\nShowing information for category: {categories[category]}")
    for i, employee in enumerate(employees):
        print(f"Employee {i+1} {categories[category]}: {employee[category]}")
```

```python
# conducting interviews and storing responses
print("Start the Interview process:")
interviewee1 = interview()
interviewee2 = interview()
interviewee3 = interview()
interviewee4 = interview()
interviewee5 = interview()

# creating a list of employees
employees = [interviewee1, interviewee2, interviewee3]
```

```python
# print out all categories of the employees
print("\nEmployee Data:")
for i, employee in enumerate(employees):
    print(f"Employee {i+1}: Name - {employee[0]}, Age - {employee[1]}, GPA - {employee[2]}")
```

```python
# prompt manager for category of interest
category = int(input("\nEnter 0 to access Name, 1 to access Age, 2 to access GPA: "))

# print out the managers category of interest for each of the employees
print_category(employees, category)
```
