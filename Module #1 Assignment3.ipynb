```python
import doctest


def ageCategory1(age):
    if age < 5:
        print("You belong to the age category: Less than 5")
    elif age >= 5 and age <= 10:
        print("You belong to the age category: Between 5 and 10 inclusive")
    elif age >= 11 and age <= 20:
        print("You belong to the age category: Between 11 and 20 inclusive")
    elif age >= 21 and age <= 64:
        print("You belong to the age category: Between 21 and 64 inclusive")
    else:
        print("You belong to the age category: Greater than 64")
"""
Test the following function:
#ageCategory1(age):

>>> ageCategory1(15)
You belong to the age category: Between 11 and 20 inclusive
"""
# Call the function from the main program

if __name__ == "__main__":

    age = int(input("Enter your age: "))
    ageCategory1(age)
    doctest.testmod()
```
